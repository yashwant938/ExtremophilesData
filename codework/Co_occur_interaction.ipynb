```python
import pandas as pd

def remove_duplicate_sentences(text):
    if pd.isna(text):
        return text

    # Split the text into sentences
    text = text.lower()
    sentences = text.split(";")

    # Remove duplicate sentences
    unique_sentences = list(set(sentences))
    unique_sentences = [s.strip() for s in unique_sentences if s.strip() != ""]

    # Sort the unique sentences alphabetically
    unique_sentences.sort()

    # Join the unique sentences back into a single string
    result = ";".join(unique_sentences)

    return result

# Read the CSV file
df = pd.read_csv("gene_chemicals2.csv")

# Apply the remove_duplicate_sentences function to each value in the "Sentences" column
df["Sentences"] = df["Sentences"].apply(remove_duplicate_sentences)

# Save the modified DataFrame to a new CSV file
df.to_csv("gene_chemicals_up2.csv", index=False)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("gene_chemicals_up2.csv")

# Function to filter lines containing both chemicals_x and chemicals_y terms
def filter_lines(sentences, chemicals_x, chemicals_y):
    if pd.isna(sentences):
        return ""

    filtered_lines = []
    for line in sentences.split(';'):
        if chemicals_x.lower() in line.lower() and chemicals_y.lower() in line.lower():
            filtered_lines.append(line)
    return ';'.join(filtered_lines)

# Apply the filter_lines function to each row
df['co-occur_sentences'] = df.apply(lambda row: filter_lines(row['Sentences'], row['Chemicals'], row['Gene']), axis=1)

# Remove rows with empty 'co-occur_sentences'
df = df[df['co-occur_sentences'].str.strip() != ""]

# Save the modified DataFrame to a new CSV file
df.to_csv("Gene_Chemicals_occur_sentences.csv", index=False)
```

```python
import pandas as pd
import re

# Read the CSV file with specified data types for each column
#dtype_dict = {'Chemicals_x': 'str', 'ChemicalsID_x': 'str','PMID': 'str','PMCID_x':'str','Chemicals_y': 'str', 'ChemicalsID_y': 'str','Sentences':'str','co-occur_sentences':'str'}  # Specify the data types for each column
merged_df = pd.read_csv('Gene_Chemicals_occur_sentences.csv')

selected_columns = ['Chemicals', 'ChemicalsID', 'PMID', 'PMCID_x', 'Gene', 'GeneID', 'Sentences', 'co-occur_sentences']
merged_df = merged_df[selected_columns]

merged_df['co-occur_sentences'] = merged_df['co-occur_sentences'].str.replace('\n', '')


def extract_interaction_types(sentence):
    if pd.isna(sentence) or sentence.strip() == '':
        return "NULL"

    interaction_types = ['Inhibition', 'Activation', 'Proliferation', 'Allosteric', 'Agonist', 'Antagonist']
    for interaction_type in interaction_types:
        if re.search(r'\b' + re.escape(interaction_type.lower()) + r'\b', sentence.lower(), re.IGNORECASE):
            return interaction_type
    return 'others'

# Drop rows with NaN values in the 'co-occur_sentences' column
merged_df.dropna(subset=['co-occur_sentences'], inplace=True)
merged_df = merged_df.loc[:, ~merged_df.columns.str.contains('^Unnamed')]

# Create a copy of the DataFrame
merged_df_copy = merged_df.copy()

# Apply the function to the 'co-occur_sentences' column and create a new column 'interaction_type'
merged_df_copy['interaction_type'] = merged_df_copy['co-occur_sentences'].apply(extract_interaction_types)

print(merged_df_copy)


merged_df_copy.to_csv('output_co-occurrence_chemicals_gene.csv', index=False)

print("DataFrame saved to output_co-occurrence_chemicals_gene.csv")
```
