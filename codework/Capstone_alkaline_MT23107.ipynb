```r
import pandas as pd

def filter_data(input_file, output_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(input_file, delimiter=',')

    # Check if the specified column exists in the DataFrame
    column_name = 'RightPart'
    if column_name not in df.columns:
        print(f"Column '{column_name}' not found in the DataFrame.")
        return

    # Filter out rows where "No Data" or "None" is present in the specified column
    df_filtered = df[(df[column_name] != 'No Data') & (df[column_name].notna())]

    # Write the filtered DataFrame to a new CSV file
    df_filtered.to_csv(output_file, sep=',', index=False)

# Replace 'OutputDisease.csv' with the actual file name
input_file = 'OutputDisease.csv'
output_file = 'FilteredOutputDisease.csv'

filter_data(input_file, output_file)
```

### **Task 10 we hade to clubbed the same species genes chemical and Disease**

```r
import csv

# Function to merge chemicals
def merge_chemicals(input_file, output_file):
    merged_chemicals = {}

    # Read data from input CSV file
    with open(input_file, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip header
        for row in reader:
            chemical, chemical_id, length, pmids_str = row[0], row[1], int(row[2]), row[3]
            pmids = eval(pmids_str) if pmids_str else []  # Convert string representation to list
            if chemical_id in merged_chemicals:
                merged_chemicals[chemical_id][3].extend(pmids)
                merged_chemicals[chemical_id][2] += length
                if chemical not in merged_chemicals[chemical_id][0]:
                    merged_chemicals[chemical_id][0].append(chemical)
            else:
                merged_chemicals[chemical_id] = [[chemical], chemical_id, length, pmids]

    # Write merged data to output CSV file
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Chemical", "ChemicalID", "length", "PMID"])
        for values in merged_chemicals.values():
            chemicals, chemical_id, length, pmids = values
            chemicals_str = ", ".join(chemicals)
            pmids_str = ", ".join(str(pmid) for pmid in pmids)
            writer.writerow([chemicals_str, chemical_id, length, pmids_str])

# Example usage:
input_file = '/content/Week6_OutputGroupByChemical.csv'
output_file = 'Task10_Clubbed_Species.csv'
merge_chemicals(input_file, output_file)
```

```r
import pandas as pd
import re


def preprocess_sentence(sentence):
    # Remove punctuation and convert to lowercase
    sentence = re.sub(r'[^\w\s]', '', sentence).lower()
    return sentence

# Convert 'Sentences' column to string
genes_df['Sentences'] = genes_df['Sentences'].astype(str)
chemicals_df['Sentences'] = chemicals_df['Sentences'].astype(str)

# Preprocess sentences in both dataframes
genes_df['Sentences'] = genes_df['Sentences'].apply(preprocess_sentence)
chemicals_df['Sentences'] = chemicals_df['Sentences'].apply(preprocess_sentence)

# Merge the two dataframes based on common PMCID
merged_df = pd.merge(genes_df, chemicals_df, on=['PMCID'], how='inner')

# Filter rows where one sentence is a subset of the other
merged_df['is_subset'] = merged_df.apply(lambda row: row['Sentences_x'] in row['Sentences_y'] or row['Sentences_y'] in row['Sentences_x'], axis=1)
merged_df = merged_df[merged_df['is_subset']]

# Select only the required columns for the final CSV
final_df = merged_df[['PMID_x', 'PMCID', 'Genes', 'Chemicals', 'Sentences_x', 'Sentences_y']]

# Rename columns for clarity
final_df.columns = ['PMID', 'PMCID', 'Genes', 'Chemical', 'Sentence_x', 'Sentence_y']

# Write the merged data into a new CSV file
final_df.to_csv('merged_data.csv', index=False)
```
